```python
import pandas as pd

df = pd.read_csv('book.csv')

df.head()
```

```python
def is_integer(value):
    try:
        if pd.isnull(value):
            return False
        int(value)
        return True
    except (ValueError, TypeError):
        return False
    
mask = df['pages'].apply(is_integer)

num_non_int_rows = (~mask).sum()
print(f"정수가 아닌 'pages' 값을 가진 행의 수: {num_non_int_rows}")

filtered_df = df[mask].copy()

filtered_df['pages'] = filtered_df['pages'].astype(int)
filtered_df.drop(['Unnamed: 0'],axis=1)
```
